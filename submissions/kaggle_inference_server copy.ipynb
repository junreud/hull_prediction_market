## 1Ô∏è‚É£ Setup: ÌôòÍ≤Ω ÏÑ§Ï†ï

````python
import os
import sys
from pathlib import Path

print("="*80)
print("LOCAL TRAINING ENVIRONMENT SETUP")
print("="*80)

# ÌîÑÎ°úÏ†ùÌä∏ Î£®Ìä∏ Í≤ΩÎ°ú
project_root = Path.cwd().parent if Path.cwd().name == 'submissions' else Path.cwd()

# Python pathÏóê Ï∂îÍ∞Ä
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

print(f"‚úì Project root: {project_root}")

# Îç∞Ïù¥ÌÑ∞ Í≤ΩÎ°ú ÌôïÏù∏
data_path = project_root / "data" / "raw"
train_file = data_path / "train.csv"
test_file = data_path / "test.csv"

if train_file.exists() and test_file.exists():
    print(f"‚úì Data files found:")
    print(f"  - {train_file}")
    print(f"  - {test_file}")
else:
    print(f"\n‚ö†Ô∏è  Data files not found!")
    print(f"  Expected location: {data_path}")
    print(f"  Please download train.csv and test.csv from Kaggle")

# artifacts Ìè¥Îçî ÏÉùÏÑ±
artifacts_dir = project_root / "artifacts"
(artifacts_dir / "models_optimized").mkdir(parents=True, exist_ok=True)
(artifacts_dir / "models_risk_optimized").mkdir(parents=True, exist_ok=True)

print(f"\n‚úì Artifacts directory ready: {artifacts_dir}")
print("\n‚úÖ Setup complete!")
````

## 2Ô∏è‚É£ Step 1: Train Return Model (~10 min)

````python
print("="*80)
print("STEP 1/3: TRAINING RETURN MODEL")
print("="*80)

# Return model ÌïôÏäµ Ïä§ÌÅ¨Î¶ΩÌä∏ Ïã§Ìñâ
script_path = project_root / "scripts" / "optimize_return_model.py"

if script_path.exists():
    print(f"‚ñ∂ Running: {script_path.name}\n")
    
    # ÌòÑÏû¨ ÎîîÎ†âÌÜ†Î¶¨Î•º ÌîÑÎ°úÏ†ùÌä∏ Î£®Ìä∏Î°ú Î≥ÄÍ≤Ω
    original_cwd = os.getcwd()
    os.chdir(project_root)
    
    %run {script_path}
    
    # ÏõêÎûò ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥µÍ∑Ä
    os.chdir(original_cwd)
    
    print("\n‚úÖ Return model training complete!")
    print(f"üìä Model saved to: artifacts/models_optimized/")
else:
    raise FileNotFoundError(f"Script not found: {script_path}")
````

## 3Ô∏è‚É£ Step 2: Train Risk Model (~10 min)

````python
print("="*80)
print("STEP 2/3: TRAINING RISK MODEL")
print("="*80)

# Risk model ÌïôÏäµ Ïä§ÌÅ¨Î¶ΩÌä∏ Ïã§Ìñâ
script_path = project_root / "scripts" / "optimize_risk_model.py"

if script_path.exists():
    print(f"‚ñ∂ Running: {script_path.name}\n")
    
    # ÌòÑÏû¨ ÎîîÎ†âÌÜ†Î¶¨Î•º ÌîÑÎ°úÏ†ùÌä∏ Î£®Ìä∏Î°ú Î≥ÄÍ≤Ω
    original_cwd = os.getcwd()
    os.chdir(project_root)
    
    %run {script_path}
    
    # ÏõêÎûò ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥µÍ∑Ä
    os.chdir(original_cwd)
    
    print("\n‚úÖ Risk model training complete!")
    print(f"üìä Model saved to: artifacts/models_risk_optimized/")
else:
    raise FileNotFoundError(f"Script not found: {script_path}")
````

## 4Ô∏è‚É£ Step 3: Optimize Position Strategy (~10 min)

````python
print("="*80)
print("STEP 3/3: OPTIMIZING POSITION STRATEGY")
print("="*80)

# Position strategy ÏµúÏ†ÅÌôî Ïä§ÌÅ¨Î¶ΩÌä∏ Ïã§Ìñâ
script_path = project_root / "scripts" / "optimize_position_strategy.py"

if script_path.exists():
    print(f"‚ñ∂ Running: {script_path.name}\n")
    
    # ÌòÑÏû¨ ÎîîÎ†âÌÜ†Î¶¨Î•º ÌîÑÎ°úÏ†ùÌä∏ Î£®Ìä∏Î°ú Î≥ÄÍ≤Ω
    original_cwd = os.getcwd()
    os.chdir(project_root)
    
    %run {script_path}
    
    # ÏõêÎûò ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥µÍ∑Ä
    os.chdir(original_cwd)
    
    print("\n‚úÖ Position strategy optimization complete!")
    print(f"üìä Strategy saved to: artifacts/position_strategy.json")
else:
    raise FileNotFoundError(f"Script not found: {script_path}")
````
