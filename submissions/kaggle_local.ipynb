## 1ï¸âƒ£ Setup: í™˜ê²½ ì„¤ì •

````python
import os
import sys
from pathlib import Path

# PyArrow í™•ì¥ íƒ€ì… ì¶©ëŒ ë°©ì§€
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', message='.*arrow.py_extension_type.*')
warnings.filterwarnings('ignore', message='.*pandas.period.*')

print("="*80)
print("LOCAL TRAINING ENVIRONMENT SETUP")
print("="*80)

# í”„ë¡œì íŠ¸ ë£¨íŠ¸ ê²½ë¡œ
project_root = Path.cwd().parent if Path.cwd().name == 'submissions' else Path.cwd()

# Python pathì— ì¶”ê°€
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

print(f"âœ“ Project root: {project_root}")

# ë°ì´í„° ê²½ë¡œ í™•ì¸
data_path = project_root / "data" / "raw"
train_file = data_path / "train.csv"
test_file = data_path / "test.csv"

if train_file.exists() and test_file.exists():
    print(f"âœ“ Data files found:")
    print(f"  - {train_file}")
    print(f"  - {test_file}")
else:
    print(f"\nâš ï¸  Data files not found!")
    print(f"  Expected location: {data_path}")
    print(f"  Please download train.csv and test.csv from Kaggle")

# artifacts í´ë” ìƒì„±
artifacts_dir = project_root / "artifacts"
(artifacts_dir / "models_optimized").mkdir(parents=True, exist_ok=True)
(artifacts_dir / "models_risk_optimized").mkdir(parents=True, exist_ok=True)

print(f"\nâœ“ Artifacts directory ready: {artifacts_dir}")
print("\nâœ… Setup complete!")
````
